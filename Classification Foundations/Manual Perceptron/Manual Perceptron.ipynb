## Manual Perceptron

A Manual Perceptron is a binary classifier. 

It takes in multiple inputs, runs it through weights and biases and outputs 1 or 0.

After getting the prediction, it compares it to the target and updates the weights and biases if incorrect.

### Importing Dependencies <a id = "idp">

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_blobs
```

### Defining the Manual Perceptron <a id = "dmp">

```python
class Perceptron:
    def __init__(self, learning_rate, epochs):
        self.lr = learning_rate
        self.epochs = epochs
        self.weights = None
        self.bias = None
```

```python
def activation(self, x):
    return 1 if x >= 0 else 0

Perceptron.activation = activation
```

```python
def fit(self, X, y):
    # Initialize the weights for each feature at 0
    self.weights = np.zeros(X.shape[1])
    # One neuron only needs 1 bias
    self.bias = 0
    
    # Epoch = Iterations
    for epoch in range(self.epochs):
        for idx, x_i in enumerate(X):
            # z = (w1*x1 + w2*x2 + ...) + b
            # Linear Output is the raw "score" calculated
            linear_output = np.dot(x_i, self.weights) + self.bias

            # The binary prediction
            prediction = self.activation(linear_output)

            # Will not update if y and prediction is the same: 0 and 0 / 1 and 1
            # If different, 0 and 1 / 1 and 0, the difference will be used as the "direction of update"
            # Step size will then be (learning rate x direction)            
            update = self.lr * (y[idx] - prediction)

            # Update parameters. If prediction is correct, update will be 0 and parameters will not change
            self.weights += update * x_i
            self.bias += update

Perceptron.fit = fit
```

```python
def predict(self, X):
    linear_output = np.dot(X, self.weights) + self.bias
    return np.array([self.activation(i) for i in linear_output])

Perceptron.predict = predict
```

### Testing the model <a id = "ttm">

#### Simple AND test

```python
# Define values
X_train = np.array([[0,0], [0,1], [1,0], [1,1]])
y_train = np.array([0, 0, 0, 1])

# Define model and fit
model = Perceptron(learning_rate=0.1, epochs=10)
model.fit(X_train, y_train)

# Display the final parameters and predicted values
print(f"Final Weights: {model.weights}")
print(f"Final Bias: {model.bias}")
print(f"Predictions: {model.predict(X_train)}")
```

#### Dataset with more values

```python
# Create 100 data points with 2 features
X, y = make_blobs(n_samples=100, centers=2, n_features=2, cluster_std=1.5, random_state=3)

# Split into Training and Testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Define model and fit
model = Perceptron(learning_rate=0.01, epochs=100)
model.fit(X_train, y_train)

# Get predicted values
test_predictions = model.predict(X_test)

# Calculate Accuracy
accuracy = np.mean(test_predictions == y_test)
print(f"Final Test Accuracy: {accuracy * 100:.2f}%")
```
