### Activation Functions
    
    Activation Functions are critical in allowing the model to fit complex curvy data patterns, different from the standard linear models.
    
    They help to decide which information is important enough to pass through, and how to treat the information to move to the next layer.

### Common Activation Functions <a id = "caf">
Sigmoid:
* Reason: Predicting a binary outcome
* Model: Classification
* Output Range: 0 to 1
* Best Layer: Final Output Layer (Binary)

tanh:
* Reason: Hidden layers where data needs to be centered around 0
* Model: Classification/Regression
* Output Range: -1 to 1
* Best Layer: Hidden Layers (RNNs, LSTMs)

reLU:
* Reason: Processing data in hidden layers to avoid math clogging
* Model: Classification/Regression
* Output Range: 0 to inf
* Best Layer: Hidden Layers (Standard CNNs, MLPs)

leaky reLU:
* Reason: Fixing dead neurons by allowing a tiny negative flow
* Model: Classification/Regression
* Output Range: -inf to inf (Minimised neg values)
* Best Layer: Hidden Layers (Deep Networks) 

SoftMax:
* Reason: Picking one "winner" from multiple categories
* Model: Multi-Class Classification
* Output Range: 0 to 1 (Outputs will sum to 1)
* Best Layer: Final Output Layer (Multi-Class)

SoftPlus:
* Reason: Predicting positive values that need smooth gradients
* Model: Regression
* Output Range: >0 to inf (Non-zero)
* Best Layer: Final Output Layer (Regression)

### Activation Function Formulas <a id = "aff">

```python
import numpy as np
```

```python
def sigmoid(z):
    # Output: A probability between 0 and 1
    return 1 / (1 + np.exp(-z))

def relu(z):
    # Output: z value if it is positive
    return np.maximum(0, z)

def softmax(z):
    # Output: Probability that adds up to 1.0 (100%)
    exp_z = np.exp(z - np.max(z)) # max(z) prevents overflow
    return exp_z / exp_z.sum()

def softplus(z):
    #Output: Non-zero positive number
    return np.log(1 + np.exp(z))

def tanh(z):
    #Output: A value between -1 and 1
    return (np.exp(z) - np.exp(-z)) / (np.exp(z) + np.exp(-z))

def leaky_relu(z, alpha=0.01):
    # Output: z value if positive, minimised z value if negative
    return np.where(z > 0, z, z * alpha)
```

### Showing the output <a id = "sto">

```python
test_inputs = np.array([-5.0, -1.0, 0.0, 1.0, 5.0])

print(f"{'Input (z)':>10} | {'Sigmoid':>8} | {'ReLU':>8} | {'LeakyR':>8} | {'Tanh':>8} | {'Softplus':>8}")
print("-" * 65)

sigmoid_o = sigmoid(test_inputs)
relu_o = relu(test_inputs)
leakyr_o = leaky_relu(test_inputs)
tanh_o = tanh(test_inputs)
softplus_o = softplus(test_inputs)
softmax_o = softmax(test_inputs)

for i in range(len(test_inputs)):
    print(f"{test_inputs[i]:>10.1f} | {sigmoid_o[i]:>8.3f} | {relu_o[i]:>8.3f} | {leakyr_o[i]:>8.3f} | {tanh_o[i]:>8.3f} | {softplus_o[i]:>8.3f}")

print("\nSoftmax Output:")
print(softmax_o)
print(f"Total Sum: {np.sum(softmax_o)}")
```

##### Sigmoid: 
Gives a probability/confidence level for Classification

##### ReLU: 
Removes negative numbers to keep it simple

##### LeakyR: 
Allows a minimised value of negative values through to keep neurons alive

##### Tanh: 
Stabilises the math by centering outputs near 0, processes negative values to discourage it

##### Softplus: 
Gentler version of ReLU for calculus-based optimization

##### SoftMax: 
Gives a probability value based on all inputs
