## Multi-Layer Perceptron with Backpropagation
This exercise aims to show a multi-layer perceptron with a backpropagation function to update the parameters (weights and biases).

### Importing Dependencies <a id = "id">

```python
import numpy as np
```

### Defining Multi-Layer Perceptron <a id = "dmlp">

```python
class MLP:
    def __init__(self, input_size, hidden_size, lr, epochs, output_size=1):
        self.lr = lr
        self.epochs = epochs
        self.hidden_size = hidden_size
        
        # Random initial parameters
        self.W1 = np.random.randn(input_size, hidden_size) * 0.01
        self.b1 = np.zeros((1, hidden_size))
        
        self.W2 = np.random.randn(hidden_size, output_size) * 0.01
        self.b2 = np.zeros((1, output_size))
```

```python
# Activation Functions
def relu(self, z): return np.maximum(0, z)
def relu_deriv(self, z): return (z > 0).astype(float)

def sigmoid(self, z): return 1 / (1 + np.exp(-z))
def sigmoid_deriv(self, z): 
    s = self.sigmoid(z)
    return s * (1 - s)
  
MLP.relu = relu
MLP.relu_deriv = relu_deriv
MLP.sigmoid = sigmoid
MLP.sigmoid_deriv = sigmoid_deriv
```

```python
# Forward Pass through parameters
def forward(self, X):
    # Hidden Layer (#1)
    self.z1 = np.dot(X, self.W1) + self.b1
    self.a1 = self.relu(self.z1)

    # Output Layer (#2)
    self.z2 = np.dot(self.a1, self.W2) + self.b2
    self.a2 = self.sigmoid(self.z2)
    return self.a2

MLP.forward = forward
```

```python
def fit(self, X, y):
    y = y.reshape(-1, 1)

    for epoch in range(self.epochs):
        # Forward Pass
        prediction = self.forward(X)

        # Backpropagation (Output)
        error_out = prediction - y
        delta_out = error_out * self.sigmoid_deriv(self.z2)

        # Backpropagation (Hidden Layer)
        error_hidden = np.dot(delta_out, self.W2.T)
        delta_hidden = error_hidden * self.relu_deriv(self.z1)

        # Update Parameters
        self.W2 -= self.lr * np.dot(self.a1.T, delta_out)
        self.b2 -= self.lr * np.sum(delta_out, axis=0, keepdims=True)

        self.W1 -= self.lr * np.dot(X.T, delta_hidden)
        self.b1 -= self.lr * np.sum(delta_hidden, axis=0, keepdims=True)

        if epoch % 10000 == 0:
            loss = np.mean(np.square(error_out))
            print(f"W1: {self.W1},\nb1: {self.b1}, \nW2: {self.W2}, \nb2: {self.b2}")
            print(f"Epoch {epoch}, Loss: {loss:.6f}\n")
            
MLP.fit = fit
```

```python
def predict(self, X):
    # Round the probability to 0 or 1
    return (self.forward(X) > 0.5).astype(int)

MLP.predict = predict
```

### Using the model <a id = "utm">

```python
X_xor = np.array([[0,0], [0,1], [1,0], [1,1]])
y_xor = np.array([0, 1, 1, 0])

# Hidden size of at least 2 or 4 for XOR
model = MLP(input_size=2, hidden_size=4, lr=0.1, epochs = 10000)
model.fit(X_xor, y_xor)

print("Expected Results:\n", y_xor)
print("\nFinal Results:")
print(model.predict(X_xor))
```

#### Increasing Iterations and Decreasing Learning Rate
Previous iterations and learning rate: 10000, 0.1

New iterations and learning rate: 20000, 0.05

```python
X_xor = np.array([[0,0], [0,1], [1,0], [1,1]])
y_xor = np.array([0, 1, 1, 0])

model2 = MLP(input_size=2, hidden_size=4, lr=0.05, epochs = 20000)
model2.fit(X_xor, y_xor)

print("Expected Results:\n", y_xor)
print("\nFinal Results:")
print(model2.predict(X_xor))
```
