Gradient Descent aims to find the optimal values of parameters (weights and bias) by minimzing a loss function. 
1) Initialize random parameter values

2) Calculate the error between the predictions and the actual target values

3) Obtain the derivative of the loss function with respect to the parameter

4) Update the parameter values based on the derivative. Step size = derivative x learning rate

5) Repeat steps 2-4 until step size reaches close to 0 or number of steps hits the defined maximum

### Importing Dependencies <a id = "idp">

```python
import numpy as np
```

```python
# Declaring test data
X = np.array([1, 2, 3, 4, 5])
Y = np.array([5, 7, 9, 11, 13]) # Target: y = 2x + 3
```

### Defining the Gradient Descent Function <a id = "gd">

```python
def gradient_descent(x, y, iterations, learning_rate):
    # Start with 1 weight and 1 bias
    w = 0.0
    b = 0.0
    n = len(x)
    
    for i in range(iterations):
        # Forward Pass to get initial prediction
        y_predicted = w * x + b
        
        # Get the difference between the predicted and target value
        # MSE is used in this circumstance
        cost = (1/n) * sum([val**2 for val in (y - y_predicted)])

        # Calculate the derivatives
        dw = -(2/n) * sum(x * (y - y_predicted))
        db = -(2/n) * sum(y - y_predicted)
        #Derivative of w: 
        #d(Cost)/d(w) = d(1/n * sum(y-y pred)^2) with respect to w
        #             = 1/n*(sum(2(y-y pred)) x d(y-y pred)
        #             = 1/n*(sum(2(y-y pred)) x d(y - (wx + b))
        #             = 1/n*(sum(2(y-y pred)) x d(y - wx - b)
        #             = 1/n*(sum(2(y-y pred)) x (0 - x * 0 - 0)
        #             = 1/n*(sum(2(y-y pred)) x (-x)
        #             = -2/n*(sum(x(y-y pred)))
        
        # Update parameters based on (derivative x learning rate)
        w = w - (learning_rate * dw)
        b = b - (learning_rate * db)
        
        if i % (iterations/5) == 0:
            print(f"Iteration {i}: Cost {cost:.4f}, Weight {w:.4f}, Bias {b:.4f}")
            
    return w, b
```

### Executing the function with different parameters <a id = "dp">

#### Increasing iterations while keeping at learning rate 0.01

```python
#10 Iterations, 0.01 Learning Rate
w_final, b_final = gradient_descent(X, Y, 10, 0.01)
print(f"\nFinal Result: y = {w_final:.2f}x + {b_final:.2f}")
```

```python
#100 Iterations, 0.01 Learning Rate
w_final, b_final = gradient_descent(X, Y, 100, 0.01)
print(f"\nFinal Result: y = {w_final:.2f}x + {b_final:.2f}")
```

```python
#1000 Iterations, 0.01 Learning Rate
w_final, b_final = gradient_descent(X, Y, 1000, 0.01)
print(f"\nFinal Result: y = {w_final:.2f}x + {b_final:.2f}")
```

#### Increasing Learning Rate will keeping iterations at 100

```python
#100 Iterations, 0.03 Learning Rate
w_final, b_final = gradient_descent(X, Y, 100, 0.03)
print(f"\nFinal Result: y = {w_final:.2f}x + {b_final:.2f}")
```

```python
#100 Iterations, 0.06 Learning Rate
w_final, b_final = gradient_descent(X, Y, 100, 0.06)
print(f"\nFinal Result: y = {w_final:.2f}x + {b_final:.2f}")
```

```python
#100 Iterations, 0.09 Learning Rate
w_final, b_final = gradient_descent(X, Y, 100, 0.09)
print(f"\nFinal Result: y = {w_final:.2f}x + {b_final:.2f}")
```

Shows that too high a learning rate will cause the numbers to "ping pong" back and forth at increasing step sizes, getting further away from the target.
